# Testing of YOLOv11x Model on VisDrone and NOMAD Images

```python
import cv2
import matplotlib.pyplot as plt
import os
from ultralytics import YOLO

# Paths to the test image, annotation file, and trained model
image_path = r"C:\Users\jmdgo\Downloads\VisDrone2019-DET-test-dev\images\9999979_00000_d_0000036.jpg"
annotation_path = r"C:\Users\jmdgo\Downloads\VisDrone2019-DET-test-dev\annotations\9999979_00000_d_0000036.txt"
model_path = r"C:\Users\jmdgo\Downloads\visocc_yolo v11x.pt"  # Update if needed

# Load the image for ground truth
image_gt = cv2.imread(image_path)
if image_gt is None:
    raise ValueError("Could not load the image. Check the path.")

# Read and parse the annotation file
if not os.path.exists(annotation_path):
    raise FileNotFoundError(f"Annotation file not found: {annotation_path}")

with open(annotation_path, "r") as f:
    lines = f.readlines()

# Draw ground truth bounding boxes (in BLUE)
for line in lines:
    values = list(map(float, line.strip().split(",")))  # VisDrone format is comma-separated
    x, y, w, h = map(int, values[:4])  # Extract x, y, width, height
    x1, y1, x2, y2 = x, y, x + w, y + h
    cv2.rectangle(image_gt, (x1, y1), (x2, y2), (255, 0, 0), 2)  # Blue boxes

# Convert to RGB for display
image_gt = cv2.cvtColor(image_gt, cv2.COLOR_BGR2RGB)

# Load the trained YOLOv8 model
model = YOLO(model_path)

# Run inference on the image
results = model.predict(source=image_path, show=False)

# Load the image for model predictions
image_pred = cv2.imread(image_path)
if image_pred is None:
    raise ValueError("Could not load the image. Check the path.")

# Draw YOLO predictions (in GREEN)
for box in results[0].boxes:
    coords = box.xyxy[0].cpu().numpy()  # Convert tensor to numpy array
    confidence = box.conf[0].item()  # Extract confidence score
    x1, y1, x2, y2 = map(int, coords)  # Convert coordinates to integers
    
    cv2.rectangle(image_pred, (x1, y1), (x2, y2), (0, 255, 0), 2)  # Green boxes
    label = f"{confidence:.2f}"  # Confidence score
    cv2.putText(image_pred, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# Convert to RGB for display
image_pred = cv2.cvtColor(image_pred, cv2.COLOR_BGR2RGB)

# Plot both images side by side
plt.figure(figsize=(15, 7))

# Left: Ground truth annotations
plt.subplot(1, 2, 1)
plt.imshow(image_gt)
plt.axis("off")
plt.title("Ground Truth Annotations")

# Right: YOLO model predictions
plt.subplot(1, 2, 2)
plt.imshow(image_pred)
plt.axis("off")
plt.title("YOLO Predictions")

plt.show()
```

```python
import os
from tqdm import tqdm

# Define the exact label directories
label_dirs = {
    "train": r"C:/Users/jmdgo/Downloads/VisDrone2019-DET-train/labels",
    "val": r"C:/Users/jmdgo/Downloads/VisDrone2019-DET-val/labels",
    "test-dev": r"C:/Users/jmdgo/Downloads/VisDrone2019-DET-test-dev/labels"
}

# Process each dataset split
for split, label_dir in label_dirs.items():
    new_label_dir = label_dir.replace("labels", "labels-yolo")  # New labels directory

    # Ensure the new directory exists
    os.makedirs(new_label_dir, exist_ok=True)

    print(f"Processing {split} labels...")

    for file in tqdm(os.listdir(label_dir)):
        if not file.endswith(".txt"):
            continue
        
        label_path = os.path.join(label_dir, file)
        new_label_path = os.path.join(new_label_dir, file)

        with open(label_path, "r") as f:
            lines = f.readlines()

        with open(new_label_path, "w") as f:
            for line in lines:
                parts = line.strip().split()
                if len(parts) == 6:  # Ensure we remove the last column
                    parts = parts[:-1]
                f.write(" ".join(parts) + "\n")

    print(f"✅ {split} labels fixed and saved in 'labels-yolo'!")

print("🎯 All labels have been processed successfully!")
```

```python
import time
from ultralytics import YOLO

# Paths to models and dataset config
model_paths = {
    "YOLOv8x": r"C:\Users\jmdgo\Downloads\visocc.pt",
    "YOLOv11x": r"C:\Users\jmdgo\Downloads\visocc_yolo v11x.pt"
}
dataset_yaml = r"C:\Users\jmdgo\Documents\CV\data.yaml"  # Update if needed

# Function to evaluate the model on the test dataset
def evaluate_model(model_name, model_path):
    print(f"\n🔹 Evaluating {model_name} on the test dataset...\n")

    # Load model
    model = YOLO(model_path)

    # Run validation
    start_time = time.time()
    metrics = model.val(data=dataset_yaml, split="test")
    end_time = time.time()

    # Extract correct metrics
    precision = metrics.box.precision  # Mean Precision
    recall = metrics.box.recall  # Mean Recall
    map50 = metrics.box.map50  # mAP@50
    map5095 = metrics.box.map  # mAP@[50:95]

    # Compute additional metrics
    f1_score = (2 * precision * recall) / (precision + recall + 1e-6)  # Avoid division by zero
    fpr = 1 - precision  # False Positive Rate
    fnr = 1 - recall  # False Negative Rate
    inference_time = round(end_time - start_time, 3)
    fps = round(len(metrics.box) / inference_time, 2) if inference_time > 0 else 0

    results = {
        "Model": model_name,
        "mAP@50": round(map50, 4),
        "mAP@50:95": round(map5095, 4),
        "Precision": round(precision, 4),
        "Recall": round(recall, 4),
        "F1-score": round(f1_score, 4),
        "Inference Time (s)": inference_time,
        "FPS": fps,
        "False Positive Rate (FPR)": round(fpr, 4),
        "False Negative Rate (FNR)": round(fnr, 4)
    }

    # Print results
    for key, value in results.items():
        print(f"{key}: {value}")

    return results

# Evaluate both models on the full test dataset
results_v8x = evaluate_model("YOLOv8x", model_paths["YOLOv8x"])
results_v11x = evaluate_model("YOLOv11x", model_paths["YOLOv11x"])

# Compare models
print("\n🔹 **Comparison of YOLO Models on Test Dataset:**")
for key in results_v8x.keys():
    if key != "Model":
        print(f"{key}: YOLOv8x = {results_v8x[key]}, YOLOv11x = {results_v11x[key]}")
```
